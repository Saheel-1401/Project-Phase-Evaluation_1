```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

min_mse = float('inf')
best_rs = 0

for i in range(1, 1000):
    x_train, x_test, y_train, y_test = train_test_split(xav, yav, test_size=0.20, random_state=i)
    lr = LinearRegression()
    lr.fit(x_train, y_train)
    pred = lr.predict(x_test)
    mse = mean_squared_error(y_test, pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, pred)
    
    if mse < min_mse:
        min_mse = mse
        best_rs = i
print('Best MSE:', min_mse, 'at random_state', best_rs)
print("MSE:", mse,'at random_state', best_rs)
print("RMSE:", rmse,'at random_state', best_rs)
print("R-squared:", r2,'at random_state', best_rs)
```

```python
from sklearn.model_selection import GridSearchCV

gb = GradientBoostingRegressor()
param_grid = {
    'learning_rate': [0.1, 0.05, 0.01],
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5]
}

grid_search = GridSearchCV(gb, param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(xav, yav)

best_params = grid_search.best_params_
best_mse = -grid_search.best_score_

print("Best Hyperparameters:", best_params)
print("Best MSE:", best_mse)
```

```python
# Initialize the GradientBoostingRegressor with the best hyperparameters
best_gb = GradientBoostingRegressor(learning_rate=0.01, max_depth=5, n_estimators=100)

# Retrain the model on the full training dataset
best_gb.fit(x_train, y_train)

y_pred = best_gb.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
print("MSE on Test Set:", mse)
```
